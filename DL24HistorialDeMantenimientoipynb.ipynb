# **Visualizar el historial de entrenamiento**

```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import numpy
```

### Cargar los datos

```python
datos = numpy.loadtxt("ClassifiedData.txt",delimiter=",")
```

Separar las entradas(X) y salidas (Y)

```python
X = datos[:,0:11]
Y = datos[:,11]
```

### Crear el modelo

```python
modelo = Sequential()
modelo.add(Dense(12, input_dim=11, activation='relu'))
modelo.add(Dense(8, activation='relu'))
modelo.add(Dense(1, activation='sigmoid'))
```

### Compilar el modelo

```python
modelo.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

### Entrenar el modelo

```python
historia = modelo.fit(X, Y, validation_split=0.33, epochs=150, batch_size=10, verbose=0)
```

### Mostrar las claves del objeto history de la variable historia

```python
print(historia.history.keys())
```

### Graficar la historia para accuracy (exactitud)

```python
plt.plot(historia.history['accuracy'])
plt.plot(historia.history['val_accuracy'])
plt.title('Accuracy del modelo')
plt.ylabel('Accuracy')
plt.xlabel('Época')
plt.legend(['entrenamiento','validación'],loc='upper left')
plt.show()
```

### Graficar la historia para perdida (loss)

```python
plt.plot(historia.history['loss'])
plt.plot(historia.history['val_loss'])
plt.title('Perdida del modelo')
plt.ylabel('Loss')
plt.xlabel('Época')
plt.legend(['entrenamiento','validación'],loc='upper left')
plt.show()
```
